# Combine Two data sets

```python
df1 = pd.DataFrame({'Fruits': ['Mango', 'Banana',
                               'Grapes', 'Apple',
                               'Orange'],
                    'Price':  [60, 40, 75, 100, 65]})

df2 = pd.DataFrame({'Fruits': ['Apple', 'Orange',
                               'Papaya',
                               'Pineapple', 'Mango', ],
                    'Price':  [120, 60, 30, 70, 50]})
print(df1)
print(df2)
```

```python
#Left/right/inner/outer Join
print(pd.merge(df1, df2, how='left', on='Fruits'))
```

```python
# Row-wise Concatenation df1 and df2
df3 = pd.concat([df1, df2], axis=0, ignore_index=True)
print(df3)
```

```python
#Column-wise Concatenation
data= {'DISCOUNT': [5, 7, 10, 8, 6]}
discount=pd.DataFrame(data)

d = pd.concat([df2, discount], axis=1)
print(d)
```
