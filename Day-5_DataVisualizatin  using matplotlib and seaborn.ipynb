# Matplotlib Basics

```python
#Importing Matplotlib
import matplotlib.pyplot as plt
```

```python
#Simple Line Plot

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)   # line plot
plt.show()
```

```python
#Adding Labels, Title

plt.plot(x, y)
plt.xlabel("X-axis")       # X label
plt.ylabel("Y-axis")       # Y label
plt.title("Simple Line Graph")  # Title
plt.show()
```

```python
#Customizing Line
plt.plot(x, y, color="red", marker="s", linestyle="--", linewidth=2)
plt.show()
```

```python
#Multiple Lines

y2 = [1, 4, 9, 16, 25]
plt.plot(x, y, label="y=2x")
plt.plot(x, y2, label="y=x")
plt.legend()   # show labels
plt.show()
```

```python
#Figure Size
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.show()
```

```python
#Saving Plot
plt.plot(x, y)
plt.savefig("myplot.png")   # saves as image
plt.show()
```

```python
#Line Plot
plt.plot(x, y, marker='o', color='b', linestyle='-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
#Scatter Plot
plt.scatter(x, y, color='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
#Bar Chart
categories = ["A", "B", "C", "D", "E"]
values = [4, 7, 2, 8, 6]

plt.bar(categories, values, color='orange')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()
```

```python
#Histogram
data = [1,2,2,3,3,3,4,4,5,5,5,5,6,7,8,9,9,9]

plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
#Pie Chart
sizes = [20, 30, 25, 25]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.show()
```

```python
# Example dataset
data = [7, 8, 5, 6, 9, 12, 15, 14, 6, 8, 7, 30]  # outlier = 30

plt.boxplot(data)
plt.title("Box Plot (Outlier Detection)")
plt.ylabel("Value")
plt.show()
```

```python
#Multiple Plots  plt.subplot()
 #Use plt.subplot(rows, cols, index)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]

# Create 1 row, 2 columns  plot 1
plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r-o')
plt.title("y = x")
plt.grid(True)

# plot 2
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'b--s')
plt.title("y = 2x")
plt.grid(True)

plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 4))

# First plot
axes[0].plot(x, y1, 'g-o')
axes[0].set_title("y = x")

# Second plot
axes[1].plot(x, y2, 'm--s')
axes[1].set_title("y = 2x")

# Second plot
axes[2].plot(x, y2, 'm--s')
axes[2].set_title("y = 2x")


plt.show()
```
