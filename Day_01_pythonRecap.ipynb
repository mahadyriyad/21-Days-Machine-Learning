# Dictionary & For Loop

Iterating through dictionary

```python
Student={
    'name': 'Karim',
    'age':29,
    'subject':'CSE'

}
for key, value in Student.items():
  print(key,":",value)
```

iterating nested Dictionary

```python
student_info={
    'student1':{'name':'karim','age':28},

}


for outer_key,values in student_info.items():
  print(outer_key)
  for inner_key,inner_value in values.items():
    print(inner_key,":",inner_value)
```

# Conditional Statement

If-Elif-ELse

```python
temp=23

if temp > 25:
  print(" it's hot day")

elif temp>=20:
  print("it's normal day")

else:
  print("it's cool day")
```

ternary operator

```python
temp=int(input("enter temperature"))
weather = " it's hot day" if temp > 25 else "it's normal day" if temp>=20 else "it's cool day"

print(weather)
```

Nested if-else

```python
temp=26
hum=70

if temp >25:
  if hum >60:
    print("temp+hum")
  else:
    print("its hot day only")

else:
  print("cold")
```

*recursion*

```python
# first n number summation
def sum_n(n):
  if n==0:
    return 0
  return n + sum_n(n-1)

sum_n(3)
```

# Class

```python
class Student:

    total_Student=0; # class attribute


    def __init__(self,name,id):
      self.name=name #instance attribute
      self.id=id


      Student.total_Student +=1

#----------instance method-----------#

    def display_info(self):
      print(f"Name:{self.name}, ID:{self.id}")    #obj_name.displayinfo()

#-------Class Method---------#

    @classmethod
    def count_total_student(cls):                 #class_name.count_total_student()
      return cls.total_Student


#-------Static method--------#


    @staticmethod
    def display_institution_name():
      print("UFTB")

student1=Student("kalam",21)
student2=Student("karim",26)

student1.display_info() #instance method
student2.display_info()

print(Student.count_total_student()) #class method

student1.display_institution_name() #static method
Student.display_institution_name()
```
