# Combining and Splitting Commands

```python
arr1 = np.array([[1, 2],[3, 4]])
arr2 = np.array([[5, 6],[7, 8]])

#Combine row-wise → axis=0
print(np.concatenate((arr1, arr2), axis=0))

#Combine column-wise → axis=1
print(np.concatenate((arr1, arr2), axis=1))

#Combine into 1D array → axis=None
print(np.concatenate((arr1, arr2), axis=None))

#Using np.stack() (adds a new dimension)
print(np.stack((arr1, arr2), axis=1))
```

```python
a = np.arange(9).reshape(3, 3)

#Split horizontally (by columns) → axis=1
print(np.split(a, 3, axis=1))  #Using np.hsplit(a,3) (horizontal split shortcut)

#Split vertically (by rows) → axis=0
print(np.split(a, 3, axis=0))  #Using np.vsplit(a,3) (vertical split shortcut)
```

# Indexing, Slicing and Subsetting

```python
# indexing integer
import numpy as np
a = np.array([[1, 2],
              [3, 4],
              [5, 6]])

print(a[[0, 1, 2], [0, 0, 1]])  #(0,0) → 1, (1,0) → 3, (2,1) → 6

# Boolean Indexing
a = np.array([10, 40, 80, 50, 100])
print(a[a > 50])
```

```python
#Slicing
a = np.array([1, 2, 3, 4, 7])
print(a[1:])
```

```python
#Subsetting

arr = np.array([1, 2, 3, 4, 7])

print("Original:", arr)
print("Subsetting:", arr[np.array([1, 3, -3])])
```

# Copying and Viewing Arrays

```python
#Normal Assignment (=) → Same reference
#both point to same array

arr = np.array([1, 2, 3, 4])
nc = arr   # no new memory created

print("IDs same?", id(arr) == id(nc))

nc[0] = 12
print("Original:", arr)
print("Assigned:", nc)
```

```python
#Deep Copy (.copy()) → New memory
#changing original does not affect copy
arr = np.array([1, 2, 3, 4])
c = arr.copy()

print("IDs same?", id(arr) == id(c))

#arr[0] = 12
c[0]=12
print("Original:", arr)
print("Copy    :", c)
```

```python
#View (.view()) → Different object, same data
#different IDs, but changes reflect in both because they share data

arr = np.array([1, 2, 3, 4])
v = arr.view()

print("IDs same?", id(arr) == id(v))

arr[0] = 12
print("Original:", arr)
print("View    :", v)
```

# NumPy Array Mathematics

```python
#Arithmetic Operations
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([[3, 2, 1],
              [1, 2, 3]])

np.subtract(a,b)
np.add(a,b)
np.divide(a,b)
np.multiply(a,b)
np.exp(b)
np.sqrt(b)
np.sin(a)
np.cos(b)
np.log(a)

```

```python
#Comparison Operations
print(a == b)      # elementwise equal
print(a < 2)       # elementwise less than 2
print(np.array_equal(a,b))  # check whole array equality
```

```python
#Aggregate Functions

a = np.array([[1, 2, 3],
              [4, 5, 6]])
b = np.array([[3, 2, 1],
              [1, 2, 3]])


a.sum()
a.min()
b.max(axis=0)
b.cumsum(axis=1)
a.mean()
np.median(b)
np.corrcoef(a.flatten(), b.flatten())
np.std(b)


```
